```python
import os
import csv
from docx import Document

# Path to the folder containing the .docx files
folder_path = "C:\\Users\\Sahal Saeed\\Documents\\5semester\\ML_project\\ML-Proj-Dataset\\ML-Proj-Dataset\\Train"
train_csv = "train.csv"

def extract_text_from_docx(file_path):
    """Extract text from a .docx file."""
    doc = Document(file_path)
    text = [paragraph.text.strip() for paragraph in doc.paragraphs if paragraph.text.strip()]
    return " ".join(text)  # Combine all paragraphs into a single string

def convert_docx_to_csv(folder_path, output_csv):
    """Convert all .docx files in a folder to a single CSV."""
    # Get all .docx files in the folder
    docx_files = [f for f in os.listdir(folder_path) if f.endswith('.docx')]
    
    with open(output_csv, mode='w', newline='', encoding='utf-8') as csv_file:
        writer = csv.writer(csv_file)
        # Write header row
        writer.writerow(["File Name", "Content"])
        
        for docx_file in docx_files:
            file_path = os.path.join(folder_path, docx_file)
            content = extract_text_from_docx(file_path)
            # Write file name and content to the CSV
            writer.writerow([docx_file, content])
    
    print(f"Conversion complete! Data saved to {train_csv}")

# Run the conversion
convert_docx_to_csv(folder_path, train_csv)
```

```python
def read_docx_table(file_path):
    """
    Reads the first table from a DOCX file and converts it into a DataFrame.
    """
    try:
        doc = Document(file_path)
        table = doc.tables[0]  # Assuming data is in the first table

        # Extract the header row
        headers = [cell.text.strip() for cell in table.rows[0].cells]

        # Extract the data
        data = []
        for row in table.rows[1:]:  # Skip the header row
            row_data = [cell.text.strip() for cell in row.cells]
            data.append(row_data)

        # Convert to DataFrame
        df = pd.DataFrame(data, columns=headers)
        return df
    except Exception as e:
        print(f"Could not process {file_path}: {e}")
        return pd.DataFrame()
```

```python
def load_all_files_in_directory(directory_path, dataset_label):
    """
    Loads all DOCX files from a directory and combines them into a single DataFrame.
    Adds a 'dataset' column to distinguish from the other group (train/test).
    """
    all_dfs = []
    for file_name in os.listdir(directory_path):
        if file_name.endswith('.docx'):
            file_path = os.path.join(directory_path, file_name)
            df = read_docx_table(file_path)
            if not df.empty:
                df['dataset'] = dataset_label  # Add a dataset column
                all_dfs.append(df)

    # Combine all DataFrames
    combined_df = pd.concat(all_dfs, ignore_index=True)
    return combined_df


# Load all training data
train_folder_path = "C:\\Users\\Sahal Saeed\\Documents\\5semester\\ML_project\\ML-Proj-Dataset\\ML-Proj-Dataset\\Train"  # Replace with actual path
train_df = load_all_files_in_directory(train_folder_path, 'train')

# Load all testing data
test_folder_path = "C:\\Users\\Sahal Saeed\\Documents\\5semester\\ML_project\\ML-Proj-Dataset\\ML-Proj-Dataset\\Test"  # Replace with actual path
test_df = load_all_files_in_directory(test_folder_path, 'test')

# Combine train and test datasets
combined_df = pd.concat([train_df, test_df], ignore_index=True)

# Inspect the combined DataFrame
print(combined_df.head())
```

```python
import pandas as pd
import json

# Load the CSV file
df = pd.read_csv('test.csv')  # Replace 'your_file.csv' with your file name

# Function to parse and normalize JSON data
def parse_and_flatten(content):
    try:
        # Replace single quotes with double quotes and parse JSON
        parsed_data = json.loads(content.replace("'", '"'))
        return pd.json_normalize(parsed_data)  # Flatten the JSON structure
    except json.JSONDecodeError:
        return None  # Handle invalid JSON

# Apply parsing and flattening function
flattened_dataframes = df['Content'].apply(parse_and_flatten)

# Concatenate all resulting DataFrames into a single DataFrame
test_df = pd.concat(flattened_dataframes.tolist(), ignore_index=True)

# Display the resulting DataFrame
(test_df)

# Save the DataFrame to a new CSV file for further use
# final_df.to_csv('processed_data.csv', index=False)
```
