```python
import pandas as pd
import numpy as np
```

```python
merged_df=pd.read_csv('Merge_data.csv')
```

```python
# Ensure both columns are in datetime format
merged_df['departure.actualTime'] = pd.to_datetime(merged_df['departure.actualTime'])
merged_df['departure.scheduledTime'] = pd.to_datetime(merged_df['departure.scheduledTime'])

# Calculate the delay in Timedelta format
merged_df['delay'] = merged_df['departure.actualTime'] - merged_df['departure.scheduledTime']

# Convert delay to minutes
merged_df['delay_in_minutes'] = merged_df['delay'].dt.total_seconds() / 60

# View the resulting dataframe
merged_df[['departure.scheduledTime', 'departure.actualTime', 'delay_in_minutes']]
```

```python
# Convert delay (timedelta) to minutes
merged_df['delay_in_minutes'] = merged_df['delay'].dt.total_seconds() / 60

# Replace negative values with 0
merged_df.loc[merged_df['delay_in_minutes'] < 0, 'delay_in_minutes'] = 0


merged_df[['departure.scheduledTime', 'departure.actualTime', 'delay_in_minutes']]
```
